```python
#import pandas library as pd
import pandas as pd  
```

```python
#Open and store txt file using pandas in a variable name file
file = pd.read_csv("conv.txt",sep=":",header=None)
file.head()
```

```python
#Store conversation list seperately and lower() it a new variable file_sep
file_sep = file[1].str.lower()
file_sep.head()
```

```python
# using replace function to remove all unwanted symbols and store back in
file_sep = file_sep.str.replace("?","") # remove ? symbol
file_sep = file_sep.str.replace(".","") # remove . pully stop
file_sep = file_sep.str.strip()         # remove right and left strip
file_sep = file_sep.str.replace(",","") # remove , 
file_sep = file_sep.str.replace("!","") # remove !
file_sep = file_sep.str.split(" ")      # split by space used to split words
```

```python
#The conversation list is ready to manipulate
file_sep.head()
```

```python
#Store name list seperate from DataFrame to a variable file_name
file_name = file[0]
file_name.head()
```

```python
# get unique dialogue speakers from the list and sort it assending order
all_names = sorted(file_name.unique())
all_names
```

```python
#Total number of unique dialogue speakers
len(all_names)
```

```python
# created a pandas DataFrame and store the list back in variable df
df = pd.DataFrame()
df["names"] = file_name   # name list for a dialogue speakers
df["conv"] = file_sep     # conversation list already we cleaned it
df.head()                 # DataFrame df
```

```python
#set index names in the list and drop old positional index True and then inplace True
df.set_index("names", drop=True , inplace=True)
df.head()
```

```python
#created a own function u_word_count() for counting and storing words to the dictionary = u_word_count
def u_word_count(name):                                  # in name field mention file.iloc
    global u_word_count                                  # make it global variable
    u_word_count=dict({})                                # created the dictionary storage variable
    for i in name:                                       # for loop created for index in table
        for n in i:                                      # for loop created for list in index
            for x in n:                                  # for loop for words in list
                if x not in u_word_count:                #if is used to store words in a dictionary 
                    y = 1
                    u_word_count.update({x : y})    #here x is a index and y is the value of 1 default
                else:
                    a = u_word_count[x]             #else is used to count words in dictionary
                    u_word_count.update({x:a+1})    # update used to update dictionary
    print(u_word_count)       # print function is used to show words and counts stored in dictionary
```

```python
# using created function finding out the words and counting
u_word_count(df.iloc)
```

```python
# sort the words ascending order and store to a variable
w_list_sort = dict(sorted(u_word_count.items()))
w_list_sort
```

```python
# creating my own function to find unique word among the conversation
global u_word_list                             #Make it global variable to store iterated data
u_word_list = []                               #creating a empty list in the variable
a = u_word_list                                #shorting the list variable name to a
for k in all_names:                            #for loop created for names in variable 
    for i in df.conv[k]:                       #for loop created for index in DataFrame
        for w in i:                            # for loop created for words in index         
                try:                        # try exception command used to pull over the index errors
                    if w_list_sort[w] == 1: #this statement checking words in a dictionary and values
                        a.append([k,w])     # append function for adding values to the list
                        print(k,w)          #print function for print names and unique words 
                        break               #break used to end this loop for added words in a line
                except:
                    continue
```

```python
#finding length of unique words created 
len(u_word_list)
```

```python
#create the dataframe for the list 
df = pd.DataFrame(u_word_list)
df.head()
```

```python
#set index and drop positional index and inplace set to true
df.set_index(0,drop=True, inplace=True)
df.head()
```

```python
nam = df.index.unique()  #save unique speakers in variable nam
dff = df.loc              #i save df.loc as a new variable dff
```

```python
#This function for creating a new text file by the name of the dialogue speaker
#store the unique words spoken by that character in the respective text file. 
def savefile(spk,a):
        save_txt = open(spk+".txt",'w+')
        save_txt.write(str(a))
        save_txt.close()
```

```python
# created a function
for i in nam:           #for loop created for speakers names in nam
    a = dff[i].values   #store the value in DataFrame to the variable a
    print(i,a)          #Print the name and value
    savefile(i,a)       #Save file function is working back to save a file in the name of speakers
                        #Store unique words into the file as printed below
```

```python
def open_file(spk):
        read_txt = open(spk+".txt",'r')
        a = read_txt.read()
        print(spk,a)
```
